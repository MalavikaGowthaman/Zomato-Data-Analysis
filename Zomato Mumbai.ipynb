# Cleaning the dataset 

```python
dataset.isna().sum()
```

```python
# check the redundant rows of data

wrg_data = dataset['VOTES'] == 'VOTES'
dataset[wrg_data]
```

```python
# Performing Negation of the wrong dataset and then storing the correct data back in the dataset DataFrame
#This permamnently remove the wrong data from the original dataframe
dataset = dataset[~wrg_data]
dataset
```

```python
# dropping the unwanted columns
dataset.drop(['URL','PAGE NO','CITY'],axis = 1, inplace = True)
dataset
```

```python
dataset.columns
```

```python
# removing null values
dataset.isnull().sum()
```

```python
dataset[dataset['PRICE'].isnull()]
```

```python
dataset.drop(labels=15080, axis =0, inplace = True)
```

```python
# replace the other null values with NA 

dataset.fillna('NA',inplace =True)
dataset
```

```python
dataset.isnull().sum()
```

# Replace the categorical values from numeric columns.

```python
dataset['RATING'].value_counts()
```

# 1st Way to replace values

```python
dataset['RATING'].replace(to_replace=['-','NEW','Opening'], value=['0','1','2'], inplace = True)
```

```python
dataset['RATING'].value_counts()
```

# 2nd ways to replace values

```python
dataset['VOTES'].value_counts()
```

```python
(dataset['VOTES']=='-').sum()
```

```python
replace_values ={'-':'0','NEW':'1','Opening':'2'}
dataset['VOTES'].replace(replace_values,inplace =True)
dataset
```

```python
(dataset['VOTES']=='-').sum()
```

# Changing the datatype of numeric columns to numeic datatype.

```python
dataset.info()
```

```python
dataset['PRICE'] = dataset['PRICE'].astype('int')
dataset['RATING'] = dataset['RATING'].astype('float')
dataset['VOTES'] = dataset['VOTES'].astype('int')
```

```python
dataset.info()
```

# Working with time column

```python
dataset['TIMING'].value_counts()
```

```python
# spliting the Timing column into two

time_df = dataset['TIMING'].str.split("(",n = 1,expand = True)
time_df
```

```python
# assign the above time_df to dataste
dataset['TIMING'] = time_df[0]
dataset['DAYS_OPEN']= time_df[1]
dataset.head()
```

```python
# to remove the character symbols from the days columns
dataset['DAYS_OPEN'] = dataset['DAYS_OPEN'].str.replace(')','')
dataset.head()
```

```python
# checking the null values
dataset.isna().sum()
```

```python
# fill the null values of day open with NA
dataset['DAYS_OPEN'].fillna('NA', inplace =True)
```

```python
dataset['DAYS_OPEN'].isna().sum()
```

```python
# to check the information of all columns
dataset.info()
```

# Removing the useless data 

```python
# finding those restaurant whose having rating = 0 and voting  = 0 and removing that values
useless_data = (dataset['RATING'] == 0.0)|(dataset['VOTES'] == 0)
dataset[useless_data]
```

```python
# this will remove the useless data from dataset
dataset = dataset[~useless_data]
dataset
```

# Working with Rating type column

```python
# checking the unique values of column
dataset['RATING_TYPE'].value_counts()
```

```python
# translating the texts into proper english text
dataset['RATING_TYPE'].replace(to_replace =['Veľmi dobré','Bardzo dobrze','Dobrze','Muy Bueno'], value ='Very Good', inplace = True)
dataset['RATING_TYPE'].replace(to_replace =['Skvělá volba','Muito Bom','Dobré','Priemer','Bom'], value ='Good', inplace = True)
dataset['RATING_TYPE'].replace(to_replace =['Promedio','Průměr','Buono','İyi','Bueno','Ortalama'], value ='Average', inplace = True)
dataset['RATING_TYPE'].replace(to_replace =['Velmi dobré','Çok iyi','Baik','Média','Skvělé','Biasa'], value ='Poor', inplace = True)
dataset['RATING_TYPE'].replace(to_replace =['Excelente'], value ='Excellent', inplace = True)
dataset['RATING_TYPE'].replace(to_replace =['Sangat Baik','Media'], value ='NA', inplace = True)
```

```python
# checking the unique values of column
dataset['RATING_TYPE'].value_counts()
```

# Working on Region column

```python
dataset['REGION'].value_counts()
```

```python
# need to get only region name
# going to remove the extra by regex --> regular expression
# below code will remove the words/sentence after/before .,+,--
dataset['REGION'] = dataset['REGION'].str.replace('[a-zA-Z].+-- ','',regex = True)
```

```python
# removing the west and east from the region column
dataset['REGION'] = dataset['REGION'].str.replace('West|East|west|east','',regex = True)
```

```python
dataset['REGION'].value_counts()
```

```python
# to shorten the string length

dataset['REGION'] = dataset['REGION'].str.replace('Girgaon Chowpatty','Chowpatty',regex = True)
```

```python
dataset['REGION'] = dataset['REGION'].str.replace('Mira Road','Mira',regex = True)
dataset["REGION"].value_counts()
```

# Removing all duplicate records

```python
# finding all the duplicate rows
dataset[dataset.duplicated()]
```

```python
# dropping the duplicate values
dataset= dataset.drop_duplicates()
dataset
```

## 9)- Find the list of Affordable Restaurants?

```python
max_price = zomato_df['PRICE'].max()
print(max_price)
one_fourth_price = max_price/4
one_fourth_price
```

```python
aff_res_df = zomato_df[zomato_df['PRICE']<=1250]
aff_res_df.sort_values(by='PRICE', inplace=True)
aff_res_df=aff_res_df[aff_res_df['PRICE']!=0]
aff_res_df
```

```python
highrate_aff_df = zomato_df[zomato_df['RATING']>4.5]
highrate_aff_df.sort_values(by ='PRICE', inplace= True)
highrate_aff_df
```

```python
highrate_aff_df = pd.merge(aff_res_df, highrate_aff_df, how='inner', on=['NAME', 'REGION'])
highrate_aff_df = highrate_aff_df[['NAME', 'PRICE_x', 'CUSINE_CATEGORY_x', 'REGION', 'CUSINE TYPE_x']]
highrate_aff_df.rename(columns={'NAME':'NAME', 'PRICE_x':'PRICE', 'CUSINE_CATEGORY_x':'CUSINE_CATEGORY',
'REGION':'REGION', 'CUSINE TYPE_x':'CUSINE TYPE'},inplace=True)
highrate_aff_df
```

## 10) - Find the list of most Reliable Restaurants?

```python
mean_votes = zomato_df['VOTES'].mean()
mean_votes
```

```python
mean_votes_df = zomato_df[['NAME','PRICE','CUSINE_CATEGORY','REGION','CUSINE TYPE','VOTES']]
mean_votes_df = mean_votes_df[mean_votes_df['VOTES']>162]
mean_votes_df.sort_values(by ='VOTES',inplace =True)
mean_votes_df
```

This dataframe obtained below shows the restaurants whose:
-- Cost is below 1250
-- Rating is above 4.5

```python
reliable_rest_df = pd.merge(mean_votes_df, highrate_aff_df, how = 'inner', on = ['NAME','REGION'])
reliable_rest_df = reliable_rest_df[['NAME','PRICE_x','CUSINE_CATEGORY_x','REGION','CUSINE TYPE_x']]
reliable_rest_df.rename(columns={'PRICE_x':'PRICE','CUSINE_CATEGORY_x':'CUSINE_CATEGORY','CUSINE TYPE_x':'CUSINE TYPE'}, inplace = True)
reliable_rest_df.sort_values(by ='PRICE')
```
